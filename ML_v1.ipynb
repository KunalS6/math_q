```python
from nltk.tokenize import word_tokenize
dataset_2['w_tokens'] = dataset_2['Question'].apply(word_tokenize)
```

```python
from nltk.stem import WordNetLemmatizer
lema=WordNetLemmatizer()
dataset_2['L_trans'] = dataset_2['w_tokens'].apply(lambda words: [lema.lemmatize(word) for word in words])
```

```python
dataset_2
```

```python
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
dataset_2['st_trans'] = dataset_2['L_trans'].apply(lambda x: [word for word in x if word.lower() not in stop])
```

```python
dataset_3=dataset_2.iloc[:,[0,1,-1]]
```
